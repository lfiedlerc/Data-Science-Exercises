```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(20,4), columns=['A','E','B','D'])
df
```

```python
#Sorting on index
df.sort_index(axis=1, ascending=False) #axis = 1 sorts the column labels
```

```python
#Sort on column value
df.sort_values(by='B')
```

```python
# Adding a new column through reindexing
df1 = df.reindex(index=df.index[0:4],
                columns=list(df.columns) + ['C'])

df1.loc[0:1,'C'] = 1
df1
```

```python
# Dropping rows with missing values
df1.dropna(how='any') # Returns a copy
```

```python
# Filling missing data
df1.fillna(value=5)
```

```python
# Getting a boolean mask
df1.isna()
```

```python
# Categorical data
df = pd.DataFrame({"id": [1, 2, 3, 4, 5, 6],
                   "raw_grade": ['a', 'b', 'b', 'a', 'a', 'e']})
```

```python
df['grade'] = df['raw_grade'].astype('category')
df['grade']
```

```python
# Renaming to something meaningful
print(df['grade'].cat.categories)
df['grade'].cat.categories = ['very good', 'good', 'very bad']
df['grade'].cat.categories
# Here we mapped the actual values to a category
```

```python
# Reordering and adding missing categories
df['grade'] = df['grade'].cat.set_categories(['very bad', 'bad',
                                'medium', 'good','very good'])
# Here we are defining the order on which values will be sorted if sort_values is called
# We are defining the possible categories even if not present in the data
```

```python
df.sort_values(by='grade')
```

```python
# Grouping by category
df.groupby('grade').size()
```

```python
# Plotting data
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))

ts = ts.cumsum()
ts.plot()
```

```python
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index,
                  columns=['A', 'B', 'C', 'D'])

df = df.cumsum()

plt.figure()

df.plot()

plt.legend(loc='best')
```
